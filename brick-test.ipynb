```python
import pandas as pd
import numpy as np
import re
import time
from datetime import datetime
from urllib.request import urlopen
from bs4 import BeautifulSoup
import requests
```

```python
query = 'Mobile Phone'
final = {}
name_list = []
image_list = []
price_list = []
rating_list = []

def get_data(pageNo):  
    HEADERS = ({'User-Agent':
                'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/44.0.2403.157 Safari/537.36',
                                'Accept-Language': 'en-US, en;q=0.5'})
 
    r = requests.get('http://www.amazon.in/s?' + 'k=' + query + '&page='+ str(pageNo), headers=HEADERS)
    content = r.content
    soup = BeautifulSoup(content, 'html.parser')

    
    # Finding all the elements with the prescribed class
    for d in soup.findAll('div', attrs={'class':'a-section aok-relative s-image-fixed-height'}):
        name = d.select('img')[0]['alt']
        imageSource = d.select('img')[0]['src'] 
        name_list.append(name)
        image_list.append(imageSource)
        
    for p in soup.findAll('span', attrs={'class':'a-price-whole'}):
        price_list.append(p.text)
        
    for r in soup.findAll('i', attrs={'class':'a-icon a-icon-star-small a-star-small-4 aok-align-bottom'}):
        rating = r.text
        rating_list.append(rating)
        
    # Creating dictionary for each of the categories
    final["Name"] = name_list
    final["Image"] = image_list
    final["Price"] = price_list
    final["Rating"] = rating_list

    
    return final
    

```

```python
no_pages = 5
for i in range(0, no_pages+1):
    get_data(i)
```

```python
#Converting final dictionary to dataframe
df = pd.DataFrame(dict([ (k,pd.Series(v)) for k,v in final.items() ]))
```
